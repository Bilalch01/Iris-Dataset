```python


# Get the counts of each species
species_counts = df['species'].value_counts()

# Create a pie chart
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)

# Add a title
ax.set_title('Iris Species Distribution')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Show the plot
plt.show()

# Pie chart for sepal length categories
sepal_length_bins = [0, 5, 6.5, 10]
sepal_length_labels = ['Short', 'Medium', 'Long']
df['sepal_length_category'] = pd.cut(df['sepal_length'], bins=sepal_length_bins, labels=sepal_length_labels)
sepal_length_counts = df['sepal_length_category'].value_counts()
plt.figure()
sepal_length_counts.plot.pie(autopct='%.1f%%')
plt.title('Sepal Length Categories')

# Pie chart for petal length categories
petal_length_bins = [0, 2, 4, 10]
petal_length_labels = ['Short', 'Medium', 'Long']
df['petal_length_category'] = pd.cut(df['petal_length'], bins=petal_length_bins, labels=petal_length_labels)
petal_length_counts = df['petal_length_category'].value_counts()
plt.figure()
petal_length_counts.plot.pie(autopct='%.1f%%')
plt.title('Petal Length Categories')

plt.show()
```
